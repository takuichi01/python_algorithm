```python
import random

def generate_graph(n, m):
  """ n個の頂点とm個の辺を持つグラフを作成する """
  graph_data = [[0]* n for i in range(n)]  #n*nの行列

  edge_set = set()

  while len(edge_set) < m:
    i, j = random.sample(range(n), 2)
    if i > j: i, j = j, i
    edge_set.add((i,j))
    graph_data[i][j] = graph_data[j][i] = 1
  
  return graph_data, edge_set

  
```

```python
random.seed(3)
node_num = 16
edge_num = 20
my_graph, edge_set = generate_graph(node_num, edge_num)
my_graph
```

```python
graph = nx.Graph()
for u,v in edge_set:
  graph.add_edge(u,v)


nx.draw_networkx(graph)
```

```python
#heapqの使い方

import heapq
import random

wait_heap = []
x = [random.randint(1, 10) for i in range(10)]

for i , d in enumerate(x):
  heapq.heappush(wait_heap, (d, i))

print(wait_heap)
```

```python
while wait_heap:
  p = heapq.heappop(wait_heap)
  i = p[1]
  print(p, i, sep=" ")
```

```python
while wait_heap:
  p = min(wait_heap)
  wait_heap.remove(p)
  i = p[1]
  print(p, i, sep=" ")
```
